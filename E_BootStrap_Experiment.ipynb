```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mu=10
sigma=2
size=1000
Sample = np.random.normal(mu,sigma,size)
```

```python
step=1/size
```

```python
plt.hist(Sample)
plt.show()
```

```python
jumps = [step*i for i in range(1,size+1)]
```

```python
Sample = sorted(Sample)
```

```python
plt.plot(Sample,jumps)
```
