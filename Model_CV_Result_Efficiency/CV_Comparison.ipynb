```python
import pandas as pd
import numpy as np
import os
import re
```

```python
os.listdir()
```

```python
pattern = "csv$"
name = []
n = []
ex = []
em = []
```

```python
for d in os.listdir():
    if re.search(pattern,d)!=None:
        temp = pd.read_csv(d,index_col=0)
        n.append(float(temp['Normal']))
        ex.append(float(temp['Exponential']))
        em.append(float(temp['Empirical']))
        name.append(d[0:-4])
```

```python
pd.options.display.float_format = '{:.2f}'.format
df = pd.DataFrame({'Normal':n,'Exponential':ex,'Empirical':em},index=name)
df = df.sort_values(by=['Empirical'],ascending=False)
df
```

```python
df.to_csv("MSE_all.csv")
```
