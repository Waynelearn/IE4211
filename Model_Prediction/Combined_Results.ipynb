```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
name = "Bagging"
bag = pd.read_csv(f"{name}.csv",index_col=0)
bag_sales = pd.DataFrame(bag['sales'])
bag_sales = bag_sales.rename(columns = {'sales':f"{name}_sales"})
bag_inv = pd.DataFrame(bag['inventory'])
bag_inv = bag_inv.rename(columns = {'inventory':f"{name}_inventory"})
```

```python
name1 = "DecisionTree"
dt = pd.read_csv(f"{name1}.csv",index_col=0)
dt_sales = pd.DataFrame(dt['sales'])
dt_sales = dt_sales.rename(columns = {'sales':f"{name1}_sales"})
dt_inv = pd.DataFrame(dt['inventory'])
dt_inv = dt_inv.rename(columns = {'inventory':f"{name1}_inventory"})
```

```python
name2 = "GradientBoosting"
gb = pd.read_csv(f"{name2}.csv",index_col=0)
gb_sales = pd.DataFrame(gb['sales'])
gb_sales = gb_sales.rename(columns = {'sales':f"{name2}_sales"})
gb_inv = pd.DataFrame(gb['inventory'])
gb_inv = gb_inv.rename(columns = {'inventory':f"{name2}_inventory"})
```

```python
name3 = "RandomForest"
rf = pd.read_csv(f"{name3}.csv",index_col=0)
rf_sales = pd.DataFrame(rf['sales'])
rf_sales = rf_sales.rename(columns = {'sales':f"{name3}_sales"})
rf_inv = pd.DataFrame(rf['inventory'])
rf_inv = rf_inv.rename(columns = {'inventory':f"{name3}_inventory"})
```

```python
name4 = "RandomForestQuantileRegressor"
rfr = pd.read_csv(f"{name4}.csv",index_col=0)
rfr_sales = pd.DataFrame(rfr['sales'])
rfr_sales = rfr_sales.rename(columns = {'sales':f"{name4}_sales"})
rfr_inv = pd.DataFrame(rfr['inventory'])
rfr_inv = rfr_inv.rename(columns = {'inventory':f"{name4}_inventory"})
```

```python
combined_inv = bag_inv.join([dt_inv,gb_inv,rf_inv,rfr_inv])
```

```python
combined_sales = bag_sales.join([dt_sales,gb_sales,rf_sales,rfr_sales])
```
